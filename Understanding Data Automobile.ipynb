```python
# Find outlier with IQR
for i in range(len(df.columns)):
    if (df[i].dtypes in ['int64', 'float64']):
        print('\n Attribute-', i, ':', df[i].dtypes)
        Q1 = df[i].quantile(0.25)
        print('Q1: ', Q1)
        Q3 = df[i].quantile(0.75)
        print('Q3: ', Q3)
        IQR = Q3 - Q1
        print('IQR: ', IQR)
        min = df[i].min()
        max = df[i].max()
        min_IQR = Q1 - 1.5*IQR
        max_IQR = Q3 + 1.5*IQR
        if (min < min_IQR):
            print('Low outlier is found')
        if(max > max_IQR):
            print('High outlier is found')
```
